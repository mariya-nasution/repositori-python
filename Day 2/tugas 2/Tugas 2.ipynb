# 1: Regresi Linear

Misalkan $A \times x = y$ adalah model regresi, dimana $x$ adalah vektor untuk nilai ```intercept``` dan ```slope```. Maka solusi dari persamaan regresi tersebut adalah:

$$x = (A^{T}A)^{-1}A^{T}y$$

dimana $A$ adalah matriks model dan $y$ adalah vektor response. 

1. Buatlah model regresi diatas hanya dengan menggunakan metode inverse matriks diatas
2. Bandingkan hasilnya dengan menggunakan ```method``` ```linalg.lstsq``` dari NumPy
3. Bandingkan hasilnya dengan menggunakan ```stats.linregress``` dari SciPy
4. Plot hasil regressinya

````python
def regresi_inverse_matriks(x, y):
    N = len(x)
    if N != len(y):
        print("Panjang data tidak sama")
        return
    # ubah menjadi vektor kolom
    v_x = np.reshape( variabel_x, (N,1) )
    v_y = np.reshape( variabel_y, (N,1) )
    # `A` adalah matriks $N\times2$.
    # Kolom pertama bernilai satu semua dan
    # kolom kedua berisi data `v_x`.
    A = np.matrix( np.zeros((N,2)) )
    A[:,0] = 1.0
    A[:,1] = v_x
    #
    At = A.T
    x = np.linalg.inv( At * A ) * At * v_y
    #
    return x
````

````python
x = regresi_inverse_matriks(variabel_x, variabel_y)
intercept = x[0,0]
slope = x[1,0]
print("intercept = ", intercept)
print("slope     = ", slope)
````

````python
N = len(variabel_x)
xmin = np.min(variabel_x)
xmax = np.max(variabel_x)
NptsPlot = N*5
xplt = np.linspace(xmin,xmax,NptsPlot)
yplt = slope*xplt + intercept
plt.scatter(variabel_x, variabel_y, label="data")
plt.plot(xplt, yplt, color="red", label="regresi linear")
plt.text(0.60, 3.0, r"$\beta_0$ = {:.4}".format(intercept))
plt.text(0.60, 2.5, r"$\beta_1$ = {:.4}".format(slope))
plt.legend();
````
